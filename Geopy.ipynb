```python
# import module
import geopy
import pandas as pd
# initialize Nominatim API 
geolocator = geopy.Nominatim(user_agent="my_application")

def get_state(row):
    location = geolocator.reverse((str(row['lat']), str(row['lon'])))
    state_name = location.raw['address']['state']
    return state_name
```

```python
airports = pd.read_csv(r'https://raw.githubusercontent.com/s-surya-s/DAV-5100/main/airports.csv')
test = airports.sample(10)
test['state'] = test.apply(get_state, axis=1)
test
```
